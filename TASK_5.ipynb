```python
import cv2
from google.colab.patches import cv2_imshow  # image display in Colab
```

```python
# Load the image
image_path = '/content/Screenshot 2025-02-12 090245.png'
image = cv2.imread(image_path)

# Check if the image is loaded correctly
if image is None:
    print("Error: Image not found. Check the file path:", image_path)

else:
    # Apply Gaussian Blur
    blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

```

```python
    # Apply Canny Edge Detection
    edges = cv2.Canny(blurred_image, 50, 150)
```

```python
    # Display images using cv2_imshow
    print("Original Image:")
    cv2_imshow(image)

    print("Blurred Image:")
    cv2_imshow(blurred_image)

    print("Edge Detected Image:")
    cv2_imshow(edges)
```

```python
from transformers import pipeline
```

```python
generator = pipeline("text-generation", model="gpt2")
```

```python
prompt = "Once upon a time in a futuristic city,"
```

```python
generated_text = generator(prompt, max_length=100, num_return_sequences=1)
```

```python
print(generated_text[0]["generated_text"])
```

```python
generated_texts = generator("Once upon a time in a futuristic city,",
                            max_length=100,
                            num_return_sequences=3)

for i, text in enumerate(generated_texts):
    print(f"\nGenerated Text {i+1}:\n{text['generated_text']}\n")
```

```python
generated_text = generator("Once upon a time in a futuristic city,",
                           max_length=100,
                           num_return_sequences=1,
                           temperature=0.7)  # Adjust for creativity

print(generated_text[0]["generated_text"])
```

```python
generated_text = generator("The AI assistant started acting strangely when",
                           max_length=100,
                           num_return_sequences=1,
                           top_k=50)  # Only pick from top 50 words

print(generated_text[0]["generated_text"])
```

```python
import torch

torch.manual_seed(42)  # Set random seed
generated_text = generator("Once upon a time,",
                           max_length=100,
                           num_return_sequences=1)

print(generated_text[0]["generated_text"])

```

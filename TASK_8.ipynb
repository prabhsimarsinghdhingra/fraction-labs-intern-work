```python
poems = [
    "The sun sets low, the sky turns red,\nA gentle breeze whispers words unsaid.",
    "The moonlight dances on the sea,\nA silent song, a melody free."
]
```

```python
tokenized_poems = [tokenizer(poem, padding=True, truncation=True, return_tensors="pt") for poem in poems]


```

```python
for i, tokenized in enumerate(tokenized_poems):
    print(f"Poem {i+1}:")
    print("Input IDs:", tokenized["input_ids"])
    print("Attention Mask:", tokenized["attention_mask"])
    print()

```

```python
from transformers import AutoTokenizer

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

sentences = ["The sun sets low.", "Is the sky blue?"]

tokens = tokenizer(
    sentences,
    padding="longest",        # Adjusts padding to the longest sentence
    truncation=True,          # Ensures texts fit within model limits
    max_length=12,            # Maximum token length
    return_tensors="pt",      # Returns PyTorch tensors
    add_special_tokens=True,  # Adds [CLS] and [SEP]
    return_token_type_ids=True,  # Differentiates between sentence pairs
    return_attention_mask=True,  # Helps ignore padding tokens
    return_offsets_mapping=True  # Maps tokens to original text positions
)

print("Input IDs:", tokens["input_ids"])
print("Token Type IDs:", tokens["token_type_ids"])
print("Attention Mask:", tokens["attention_mask"])
print("Offset Mapping:", tokens["offset_mapping"])
```

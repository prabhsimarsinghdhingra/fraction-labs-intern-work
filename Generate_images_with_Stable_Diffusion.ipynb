```python
# 🚀 Step 1: Install Required Libraries
!pip install diffusers transformers torch accelerate scipy safetensors

# 🚀 Step 2: Import Required Modules
import torch
from diffusers import StableDiffusionPipeline
from IPython.display import display

# 🚀 Step 3: Load Stable Diffusion Model
model_id = "runwayml/stable-diffusion-v1-5"  # You can change this to a different model
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.to("cuda")  # Move model to GPU

# 🚀 Step 4: Set Your Prompt for Image Generation
prompt = "A futuristic cyberpunk city at night, neon lights reflecting on wet streets, cinematic lighting, highly detailed"

# 🚀 Step 5: Generate the Image
image = pipe(prompt).images[0]

# 🚀 Step 6: Save & Display the Image
image.save("generated_image.png")
display(image)

# 🚀 Step 7: Download the Image (Optional)
from google.colab import files
files.download("generated_image.png")
```

```python
import requests
import pandas as pd
import time
from bs4 import BeautifulSoup
from datetime import datetime

# Configurations
TRAIN_NUMBER = "12630"  # Replace with your train number
CSV_FILE = "train_data.csv"

# Function to scrape train status
def scrape_train_status(train_number):
    URL = f"https://www.runningstatus.in/status/{train_number}"

    try:
        response = requests.get(URL, headers={"User-Agent": "Mozilla/5.0"})
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')

            # Extract train status
            status_section = soup.find("div", class_="running-status")
            if status_section:
                status_text = status_section.get_text(strip=True)
                train_data = {
                    "train_number": train_number,
                    "status": status_text,
                    "timestamp": datetime.now()
                }
                return train_data
            else:
                print("Train status not found.")
        else:
            print("Failed to fetch data. HTTP Status:", response.status_code)
    except Exception as e:
        print("Scraping Error:", e)

    return None

# Function to store data in CSV
def store_in_csv(data):
    df = pd.DataFrame([data])
    df.to_csv(CSV_FILE, mode='a', index=False, header=False)
    print("Data stored in CSV.")

# Main function to run the scraper
def main():
    data = scrape_train_status(TRAIN_NUMBER)
    if data:
        print("Train Data:", data)
        store_in_csv(data)
    else:
        print("Failed to fetch train data.")

# Run script every 10 minutes
if __name__ == "__main__":
    while True:
        main()
        time.sleep(600)  # Wait 10 minutes before running again
```

```python
import requests
import pandas as pd
import time
from bs4 import BeautifulSoup
from datetime import datetime

# Configuration
TRAIN_NUMBER = "12630"  # Replace with actual train number
CSV_FILE = "train_data.csv"
URL = f"https://www.runningstatus.in/status/{TRAIN_NUMBER}"

# Function to scrape train status using BeautifulSoup
def scrape_train_status():
    try:
        headers = {"User-Agent": "Mozilla/5.0"}
        response = requests.get(URL, headers=headers)

        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')

            # Find train running status
            status_section = soup.find("div", class_="panel-heading")
            if status_section:
                status_text = status_section.get_text(strip=True)

                train_data = {
                    "train_number": TRAIN_NUMBER,
                    "status": status_text,
                    "timestamp": datetime.now()
                }
                return train_data
            else:
                print("Train status not found.")
        else:
            print("Failed to fetch data. HTTP Status:", response.status_code)
    except Exception as e:
        print("Scraping Error:", e)

    return None

# Function to store data in CSV
def store_in_csv(data):
    df = pd.DataFrame([data])
    df.to_csv(CSV_FILE, mode='a', index=False, header=False)
    print("Data stored in CSV.")

# Main function to run the scraper
def main():
    data = scrape_train_status()
    if data:
        print("Train Data:", data)
        store_in_csv(data)
    else:
        print("Failed to fetch train data.")

# Run the script every 10 minutes
if __name__ == "__main__":
    while True:
        main()
        time.sleep(600)  # Wait 10 minutes before running again
```

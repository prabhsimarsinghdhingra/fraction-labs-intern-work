```python
from transformers import pipeline, set_seed
import gradio as gr
```

```python
generator = pipeline("text-generation", model="gpt2")
set_seed(42)
```

```python
def generate_content(prompt, content_type, max_len):
    taglines = {
        "Blog Introduction": "Here's a blog intro:\n",
        "Social Media Post": "Here's a catchy social media post:\n",
        "Product Description": "Here's a product description:\n"
    }
    pre_prompt = taglines.get(content_type, "") + prompt
    output = generator(pre_prompt, max_length=max_len, num_return_sequences=1)
    return output[0]['generated_text']
```

```python
interface = gr.Interface(
    fn=generate_content,
    inputs=[
        gr.Textbox(label="Enter your prompt here"),
        gr.Dropdown(choices=["Blog Introduction", "Social Media Post", "Product Description"], label="Content Type"),
        gr.Slider(minimum=50, maximum=200, step=10, value=100, label="Max Output Length")
    ],
    outputs="text",
    title="🧠 AI Content Generator",
    description="Generate blog intros, social posts, or product descriptions with GPT-2"
)
```

```python
interface.launch(share=True)
```

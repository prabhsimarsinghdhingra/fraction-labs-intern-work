```python
from confluent_kafka import Producer

# Kafka Configuration
conf = {'bootstrap.servers': "localhost:9092"}
producer = Producer(conf)

# Send Messages
def send_message(topic, key, value):
    producer.produce(topic, key=key, value=value)
    producer.flush()

send_message('my-topic', 'key1', 'Hello from Colab!')
print("‚úÖ Message sent successfully!")
```

```python
from confluent_kafka import Consumer

# Kafka Consumer Configuration
conf = {
    'bootstrap.servers': "localhost:9092",
    'group.id': "group1",
    'auto.offset.reset': 'earliest'
}
consumer = Consumer(conf)
consumer.subscribe(['my-topic'])

print("‚è≥ Waiting for messages...")
while True:
    msg = consumer.poll(1.0)
    if msg is None:
        continue
    print(f"üì© Received message: {msg.value().decode('utf-8')}")
    break  # Stop after receiving one message

consumer.close()
```

```python
from confluent_kafka import Producer
import time
import json
import random

# Kafka Configuration
producer = Producer({'bootstrap.servers': "localhost:9092"})

# Simulating GPS Traffic Data
gps_events = [
    {"vehicle_id": "KA-05-1234", "speed": 80, "location": "MG Road", "status": "Normal"},
    {"vehicle_id": "KA-01-5678", "speed": 20, "location": "Electronic City", "status": "Heavy Traffic"},
    {"vehicle_id": "KA-09-9876", "speed": 0, "location": "Whitefield", "status": "Accident"}
]

def send_gps_event():
    event = random.choice(gps_events)
    producer.produce('traffic-events', key=event["vehicle_id"], value=json.dumps(event))
    producer.flush()
    print(f"üì° GPS Event Sent: {event}")

# Send data every 3 seconds
while True:
    send_gps_event()
    time.sleep(3)
```

```python
cctv_events = [
    {"camera_id": "CCTV-101", "location": "MG Road", "incident": "Normal"},
    {"camera_id": "CCTV-202", "location": "Electronic City", "incident": "Heavy Traffic"},
    {"camera_id": "CCTV-303", "location": "Whitefield", "incident": "Accident"}
]

def send_cctv_event():
    event = random.choice(cctv_events)
    producer.produce('traffic-events', key=event["camera_id"], value=json.dumps(event))
    producer.flush()
    print(f"üìπ CCTV Event Sent: {event}")

# Send data every 5 seconds
while True:
    send_cctv_event()
    time.sleep(5)
```

```python
emergency_calls = [
    {"caller_id": "911-AAA", "location": "MG Road", "call_type": "Normal"},
    {"caller_id": "911-BBB", "location": "Electronic City", "call_type": "Traffic Jam"},
    {"caller_id": "911-CCC", "location": "Whitefield", "call_type": "Accident Reported"}
]

def send_emergency_call():
    event = random.choice(emergency_calls)
    producer.produce('traffic-events', key=event["caller_id"], value=json.dumps(event))
    producer.flush()
    print(f"üìû Emergency Call Sent: {event}")

# Send data every 7 seconds
while True:
    send_emergency_call()
    time.sleep(7)
```

```python
from confluent_kafka import Consumer
import json

# Kafka Consumer Configuration
consumer = Consumer({
    'bootstrap.servers': "localhost:9092",
    'group.id': "traffic-monitor",
    'auto.offset.reset': 'earliest'
})

consumer.subscribe(['traffic-events'])

print("‚è≥ Monitoring real-time traffic events...")

while True:
    msg = consumer.poll(1.0)
    if msg is None:
        continue

    event = json.loads(msg.value().decode('utf-8'))

    # Accident Detection Logic
    if event.get("status") == "Accident" or event.get("incident") == "Accident" or event.get("call_type") == "Accident Reported":
        print(f"üö® ACCIDENT DETECTED at {event.get('location')}! üöë Sending alert...")

consumer.close()
```
